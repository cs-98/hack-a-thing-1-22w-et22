```python
from panda_stats import panda_stats
from IPython.display import display
import pandas as pd
```

### Test Case 1
Run the following cell to produce the panda_stats display. At the top, you will see your input data (similar to df.head()). In the middle, you will see a series of statistical tests organized in accordions. After running a statistical test, the output will appear in the output section below. 

With this data, I recommend the following tests - 
1. Independent t-test with grouping variable `condition` and dependent variable `systolic bp`. 
2. Linear regression with independent variable `systolic bp` and dependent variable `diastolic bp`.
3. Wilcoxon signed-rank test with grouping variable `gender` and dependent variable `resting heart rate`.


```python
# test dataset 1 - bp + hr experiment
df = pd.DataFrame({"condition": pd.Series(["placebo","placebo","control","placebo","control", "control"], dtype="category"), \
                "systolic bp": [140,135,138,122, 125, 146], \
                "diastolic bp": [85,90,93,102, 100, 101], \
                "resting heart rate": [55, 60, 80, 90, 47, 59], \
                "age": pd.Series(["young", "mid", "mid", "old", "young", "old"], dtype="category"), \
                "gender": pd.Series(["M", "F", "M", "M", "F", "F"], dtype="category")})

# call pandastats
widget = panda_stats(df)
display(widget)
```
